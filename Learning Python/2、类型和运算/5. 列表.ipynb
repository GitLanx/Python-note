# 1. 列表  
列表是有序集合对象类型，可以包含任何种类的对象，且每一项都保持了从左到右的位置顺序。  

Python 列表是：
- 任意对象的有序集合
- 通过偏移读取
- 可变长度、异构以及任意嵌套
- 属于可变序列的分类
- 对象引用数组

# 2. 实际应用中的列表
## 2.1 基本列表操作

```python
len([1, 2, 3])                  # 长度
```

```python
[1, 2, 3] + [4, 5, 6]           # 合并
```

```python
['Ni!'] * 4                     # 重复
```

列表中“+”操作两边必须是相同类型的序列。

```python
[1, 2] + list('34')
```

## 2.2 列表迭代和解析

```python
3 in [1, 2, 3]              # 成员关系
```

```python
for x in [1, 2, 3]:         # 迭代
    print(x, end = '')
```

```python
[c * 4 for c in 'SPAM']     # 列表解析
```

## 2.3 索引、分片和矩阵

```python
L = ['spam', 'Spam', 'SPAM!']
print(L[2])                 # 偏移从 0 开始
print(L[-2])                # 负偏移，从右开始
print(L[1:])                
```

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1])
print(matrix[1][1])
print(matrix[2][0])
```

## 2.4 原处修改列表  
直接修改列表对象，不会建立新的拷贝。  

**索引与分片赋值**

```python
L = ['spam', 'Spam', 'SPAM!']
L[1] = 'eggs'                           #索引赋值
L
```

```python
L[0:2] = ['eat', 'more']                #分片赋值：删除+插入，替换0,1项
L
```

```python
L[1:2] = ['aaa', 'bbb']
L
```

```python
L = [1, 2, 3]
L[1:1] = [6, 7]              # 插入（什么也不替换）
L
```

```python
L[1:2] = []                  # 删除（什么也不插入）
L
```

分片赋值一次性替换整个片段或“栏”，因为被赋值的序列长度不一定要与被赋值的分片的长度相匹配，所以分片赋值能够用来替换（覆盖）、增长（插入）、缩短（删除）主列表。

```python
L = [1]
L[:0] = [2, 3, 4]                    # 在 0 处插入所有项，头部空分片
L
```

```python
L[len(L):] = [5, 6, 7]               #在 len(L) 处插入所有项，尾部空分片
L
```

```python
L.extend([8, 9, 10])                 #在尾部插入所有项，方法调用
L
```

**列表方法调用**

```python
L = ['spam', 'Spam', 'SPAM!']      # append 将一个单项加至列表末端
L.append('please')
L      
```

```python
L.sort()                           # 递增顺序排序（'S'<'e'）
L
```

```python
L = ['abc', 'ABD', 'aBe']
sorted(L, key = str.lower, reverse = True)      # 可以传入关键字参数
L
```

```python
L = [1, 2]
L.extend([3, 4, 5])        # 在末端添加多个项
L
```

```python
L.pop()                    # 删除并且返回最后一项，也能接受某一个即将删除并返回的元素的偏移
```

```python
L
```

```python
L.reverse()                # 原地反转方法
list(reversed(L))
L
```

```python
L.index(1)                 # 某项的索引
```

```python
L.insert(1,5)              # 在位置上插入
L
```

```python
L.remove(2)                # 按值删除
L
```

```python
L.pop(0)                   # 按位删除
L
```

```python
L.count(1)                 # 出现的次数
```
